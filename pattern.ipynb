```python
n=int(input())      #Rangoli
i=0
while i<n:
    j=0
    c=97
    while j<n:
        if j<((n-i)-1):
            print("-",end=' ')
        else:
            print(chr(c),end=' ')
            c+=1
        j+=1
    j=0
    c=97+(i-1)
    while j<(n-1):
        if j<i:
            print(chr(c),end=' ')
            c-=1
        else:
            print('-',end=' ')
        j+=1      
    print('')
    i+=1
while i>1:
    j=n
    c=97
    while j>0:
        if j<i:
            print(chr(c),end=' ')
            c+=1
        else:
            print('-',end=' ')
        j-=1
    j=n
    c=97+(i-3)
    while j>1:
        if j>((n-i)+2):
            print(chr(c),end=' ')
            c-=1
        else:
            print('-',end=' ')
        j-=1
    print('')
    i-=1
            
```
